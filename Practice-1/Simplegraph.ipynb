```python
from typing import TypedDict, Dict
from langgraph.graph import StateGraph
```

```python
class AgentGraph(TypedDict):
    Name:str

def compliment_generator(state:AgentGraph)->AgentGraph:
    """These is an simple node used to generate compliments"""
    print(state)
    state['Name']= state['Name'] + ",you're doing amazing job learning LangGraph"
    print(state)
    return state
```

```python
graph=StateGraph(AgentGraph)

graph.add_node("Complimentor",compliment_generator)
graph.set_entry_point("Complimentor")
graph.set_finish_point("Complimentor")

app=graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result=app.invoke({"Name":"Jayant"})
print(result)
result['Name']
```
